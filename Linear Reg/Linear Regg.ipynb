### What Are the Basic Assumption?(favourite)
There are four assumptions associated with a linear regression model:

1. Linearity: The relationship between X and the mean of Y is linear.
2. Homoscedasticity: The variance of residual is the same for any value of X.
3. Independence: Observations are independent of each other.
4. Normality: For any fixed value of X, Y is normally distributed.

### Advantages
1. Linear regression performs exceptionally well for linearly separable data
2. Easy to implement and train the model
3. It can handle overfitting using dimensionlity reduction techniques and cross validation and regularization 


### Disadvantages
1. Sometimes Lot of Feature Engineering Is required
2. If the independent features are correlated it may affect performance
3. It is often quite prone to noise and overfitting

### Overfitting And Underfitting

Linear Reg aims to create a best fit line so, most of the data can fit on this line, 
BUT if the data in curve shape, we have to change the degree of the polyniomial (simple linear is useless)

The more we increase the degree of the polynomial the shape its gets more curve

- it gets affected by outliers (bias & variance)

- If polynomial degree = 1 it means that it will act as a (Simple linear Regression) so it will try to draw a best fit line
- Sometimes we need to change the degree of the polynomial so it can get better accuracy and avoid (Under fitting) in trainig data
- however, be aware of training too much to get a very high accuracy in training set and bad accuracy in testing set which is known as (Over fitting)

#### BIAS & Variance

- BIAS = The error of (TRAINING Data)
- VARIANCE = The error of (TESTING Data)

OUT GOAL IS TO HAVE
- LOW BIAS
- LOW VARIANCE

##### Underfitting

- Training Data has low Acc (HIGH BIAS)
- Testing Data has low Acc (HIGH VARIANCE)

##### Overfitting

- Training Data has High Acc (LOW BIAS)
- Testing Data has low Acc (HIGH VARIANCE)

## Simple Linear Reg

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/homeprices.csv')
df.shape
```

```python
df[:3]
```

```python
plt.xlabel('area')
plt.ylabel('price')

plt.scatter(df['area'],df['price'],color='green',marker='+')
```

### We can see than we can create a line so we can build a simple linear reg

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop('price' , axis = 1)
y = df['price']
```

```python
reg_model = LinearRegression()

reg_model.fit(X, y)
```

```python
reg_model.coef_
```

```python
reg_model.intercept_
```

```python
y_hat = reg_model.predict(X)
```

- We can plot the line fit generated from the model

```python
plt.xlabel('area')
plt.ylabel('price')

plt.scatter(df['area'],df['price'],color='green',marker='+')

plt.scatter(df['area'],y_hat,color='Red',marker='o')
#plt.plot(df['area'], y_hat, color = 'Red')
```

```python
plt.xlabel('area')
plt.ylabel('price')

plt.scatter(df['area'],df['price'],color='green',marker='+')

#plt.scatter(df['area'],y_hat,color='Red',marker='o')
plt.plot(df['area'], y_hat, color = 'Red')
```

### Compute Accuracy

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y, y_hat)
```
